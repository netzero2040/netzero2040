# Tutorial - Using pyam for NetZero2040 scenario analysis

The project NetZero2040 will use various tools to create, analyze and visualize reference data, scenario assumptions and model results.

This tutorial notebook illustrates how the Python package [pyam](https://pyam-iamc.readthedocs.io) can be used
to facilitate the work in the project.

```python
import pyam
```

The tutorial data is taken from the Energiebilanz of *Statistik Austria*.
It shows the fuel mix for power generation over the past decade.

We first import the data, then show a simple overview of the contents.

```python
df = pyam.IamDataFrame("tutorial_data/energy-balance-austria.csv")
```

```python
df
```

This repository contains a dictionary file with plotting conventions and color schemes,
which can be used in this project to ensure a consistent visual style.

The plotting module of the **pyam** package can load this configuration file
via the [Run Control](https://pyam-iamc.readthedocs.io/en/stable/api/plotting.html) feature
and apply it easily to any figure.  
See a few examples below!

We first import the configuration file.

```python
pyam.plotting.run_control().update("styles/plotting-default.yaml")
```

The most simple type of figure is the line plot...

<div class="alert alert-warning">

Note that **pyam** will automatically choose colors or other styles for any items that are not explicitly defined.

</div>

```python
df.plot(color="variable")
```

Bar plots are also quite common in energy systems analysis...

```python
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

And as a last illustration, here is a pie plot for a particular year...

```python
df.filter(year=2019).plot.pie()
```
